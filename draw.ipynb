```python
import numpy as np 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
E = np.loadtxt("E.txt", delimiter='\t', dtype=np.float)
M = np.loadtxt("M.txt", delimiter='\t', dtype=np.float)
C = np.loadtxt("C.txt", delimiter='\t', dtype=np.float)
X = np.loadtxt("X.txt", delimiter='\t', dtype=np.float)
```

```python
f = plt.figure(figsize=(18, 10))
T = [0.1, 0.2, 0.3, 0.8]    
    
sp =  f.add_subplot(2, 2, 1 )
# plt.scatter(T, E, s=50, marker='o', color='IndianRed')
plt.plot(T, E, marker='o', color='IndianRed')
plt.xlabel("Temperature (T)", fontsize=20)
plt.ylabel("Energy ", fontsize=20)         
plt.axis('tight')


sp =  f.add_subplot(2, 2, 2 )
# plt.scatter(T, abs(M), s=50, marker='o', color='RoyalBlue')
plt.plot(T, abs(M), marker='o', color='RoyalBlue')
plt.xlabel("Temperature (T)", fontsize=20) 
plt.ylabel("Magnetization ", fontsize=20)   
plt.axis('tight')


sp =  f.add_subplot(2, 2, 3 )
# plt.scatter(T, C, s=50, marker='o', color='IndianRed')
plt.plot(T, C, marker='o', color='IndianRed')
plt.xlabel("Temperature (T)", fontsize=20)  
plt.ylabel("Specific Heat ", fontsize=20)   
plt.axis('tight')   


sp =  f.add_subplot(2, 2, 4 )
# plt.scatter(T, X, s=50, marker='o', color='RoyalBlue')
plt.plot(T, X, marker='o', color='RoyalBlue')
plt.xlabel("Temperature (T)", fontsize=20) 
plt.ylabel("Susceptibility", fontsize=20)
plt.axis('tight')   


plt.savefig("fig1.png")
```
